### Create a  barchart or histogram to visualize the distribution of a categorical or continuous variable, such as the disrubution of age or gender variables in a population.
Dataset: Health Insurance dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\91998\OneDrive\prodigy_internship_tskas\Data\insurance.csv")
df.head()
```

## Data Exploration
###Feature Description:

1. age: age of primary beneficiary
2. sex: insurance contractor gender, female, male
3. bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,   objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9
4. children: Number of dependents
5. smoker: Smoking
6. region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.
7. charges: Individual medical costs billed by health insurance

```python
df.rename(columns={'sex': 'gender'}, inplace=True)
```

```python
df.head()
```

```python
sns.histplot(df['age'], bins=10, kde=True, color='skyblue')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram Plot of Age')

# Show the plot
plt.show()
```

## This is distribution plot of age
### Observation:
- we can see that every age group has almost same distribution. But it is not normally distributed. It's almost looking symmetric but having some discrepencies in the peek of the bell shaped curve.

```python
### Vizual Analysis on gender feature using bar plot
sns.countplot(x='gender', data=df)

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Countplot of Gender')

# Show the plot
plt.show()
```

## This is the barplot of gender column
### Observation:
- Almost male and female persons are having same medical expences.

```python
sns.histplot(df['bmi'], bins=10, kde=True, color='skyblue')

# Add labels and title
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Histogram Plot of Age')

# Show the plot
plt.show()
```

## this is the distribution plot of continuous variable bmi
## Observation:
### This graph idcates that it is normally distributed. And also we can Indentify symmetricity in the bell shaped curve.
